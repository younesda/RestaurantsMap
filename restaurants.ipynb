```python
from connexion.conn_mongo import ConnexionMongo

result = ConnexionMongo.connexion_mongo()
```

```python
import pandas as pd
from transformation.convert_to_df import ConvertToDF
restaurants = ConvertToDF.convert_to_df(result)
```

```python
from package_exploration_data.doublon import Doublon
Doublon.calcul_doublons(restaurants)
from drop.drop_duplicates import DropDuplicates
restaurants = DropDuplicates.drop_duplicates(restaurants)
from drop.dropNA import DropNA
restaurants = DropNA.drop_na(restaurants)
```

```python
import folium

# Vérifier si le DataFrame n'est pas vide
if not restaurants.empty:
    avg_lat = restaurants['latitude'].mean()
    avg_lon = restaurants['longitude'].mean()
    m = folium.Map(location=[avg_lat, avg_lon], zoom_start=12)

    # Ajouter les marqueurs
    for _, row in restaurants.iterrows():
        if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
            folium.Marker(
                location=[row['latitude'], row['longitude']],
                popup=row['name'],
                icon=folium.Icon(color="blue")
            ).add_to(m)
else:
    m = folium.Map(location=[0, 0], zoom_start=2)

m.save("restaurants_map.html")
print("✅ Carte générée : restaurants_map.html")
```

```python
df = restaurants.head(20)
# Carte centrée sur la moyenne des points
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=12)

# Ajouter les marqueurs
for _, row in df.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=row['name']
        ).add_to(m)

m.save("top20_restaurants_map.html")
```
